# Colorbar

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# Define normalization and colormap
norm = mpl.colors.Normalize(vmin=-1.0, vmax=1.0)
cmap = mpl.cm.viridis
sm = mpl.cm.ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])

fig = plt.figure(figsize=(1.4, 8), layout="constrained")
ax = fig.add_subplot(111)
cbar = fig.colorbar(sm, cax=ax, orientation="vertical")
cbar.set_label("CV", fontsize=FONTSIZE_SMALL)
cbar.set_ticks([-1.0, -0.5, 0.0, 0.5, 1.0])
cbar.ax.tick_params(labelsize=FONTSIZE_SMALL)

plt.savefig(f"{img_base_dir}/colorbar.png", dpi=300)
plt.show()
```
